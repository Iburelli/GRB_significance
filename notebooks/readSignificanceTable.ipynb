```python
import numpy as np
```

```python
filename='examples/Events_1_100.North.noMoon.thresh10.hidden.npy'
```

```python
data=np.load(filename,allow_pickle=True, encoding='latin1', fix_imports=True).flat[0]
events = list(data.keys())
sites = list(data[events[0]].keys())
print(f'These are the parent keywords: \n\t{sorted(data.keys())}')
events = list(data.keys())
print(f'Inside each of the parent keyword you will find the following keys:' 
      '\n\t{sorted(data[events[0]].keys())}')
sites = list(data[events[0]].keys())
nights= data[events[0]][sites[0]].keys()
notti=[]
for night in nights:
    if night!='first_night_start':  #it's a useless paremeter, i needed it only for 
                                   #computational purposes
        notti.append(night)
        
print(f'Each key is a dictionary: \n\t{notti}')
```

```python
print(f'In example you can access data with a nested cycle:')
for n, event in enumerate(events):
    for site in sites:
        print(f'Template {event} - Site {site} ')
        if type(data[event][site]) == float:
            print(f'\tThis contains NaNs --->'
                  'the source is not observable due to daylight or moon.')
        else:
            print(f"\tSite keys are dictionaries containing details about single nights.")
            for night in data[event][site].keys():
                if night != 'first_night_start':
                    print(f"\n\t{night}:")
                    if type(data[event][site][night]['irf']) == float:
                        print(f'\tThis contains NaNs --->'
                              'the source is not observable at the site.')
                    else:
                        #print('\n')
                        print(f"\tStart time of single time intervals"
                              "(here is printed just the first value):"
                              f"\n\t{round(data[event][site][night]['t_start'][0],2)}")
                        print(f"\tStop time of single time intervals"
                              "(here is printed just the first value):"
                              f"\n\t{round(data[event][site][night]['t_stop'][0],2)}")
                        print(f"\tIrf associated to the interval:"
                              f"\n\t{data[event][site][night]['irf'][0]}")
                        print(f"\tSignificance calculated up to the interval stop time:"
                              f"\n\t{data[event][site][night]['significance'][0]}")
                        print(f"\tVariance associated to the significance:"
                              f"\n\t{data[event][site][night]['variance'][0]}")
                        print('\n')
    if n > 2:
        print("Let's stop after 3 templates.")
        break

```
